# Analyze if the API works as expected

```python
import os

import requests
import json

url = os.environ["AVIATIONSTACK_BASE_URL"] + "/flights"

resp = requests.get(
    url,
    params={
        "access_key": os.environ["AVIATIONSTACK_API_KEY"],
        "limit": os.environ["AVIATIONSTACK_LIMIT"],
        "offset": 0,
    },
    timeout=60,
)

resp.raise_for_status()

```

```python
with open("flights.json", "w") as f:
    json.dump(resp.json(), f)
```
